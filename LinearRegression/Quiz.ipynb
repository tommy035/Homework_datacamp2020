```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#12 สร้าง train/test split ของบ้าน
X = df["Avg. Area Income"]
y = df['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 80)
```

```python
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
```

```python
#13 เทรนโมเดลแบบ Simple Linear Regression 
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
predicted = lm.predict(X_test)
predicted
```

```python
#14 ทดสอบโมเดลวัดค่า MAE, MSE, RMSE
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test, predicted))
print('MSE: ', metrics.mean_squared_error(y_test, predicted))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, predicted)))
```

```python
#15 สร้าง distribution plot เพื่อดูว่า predicted results เป็น normal ไหม
fig = plt.figure(figsize=(10,8))
sns.distplot((y_test-predicted), bins=50)
```

```python
#16 สร้าง scatter plot และ prediction line ของ simple linear regression
fig = plt.figure(figsize=(10,8))
plt.scatter(X_test, y_test, color='blue', label='data')
plt.plot(X_test, lm.predict(X_test), color='red', label= 'Predicted Regression line')
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.legend()
```

```python
#17 เทรนโมเดลแบบ Multiple Linear Regression 
X2 = df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population']]
y2 = df['Price']
```

```python
X2_train,X2_test,y2_train,y2_test = train_test_split(X2,y2, test_size=0.2, random_state=80)
```

```python
lm = LinearRegression()
lm.fit(X2_train,y2_train)
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_, X2.columns,columns=['Coefficient'])
coeff_df
```

```python
#18 ทดสอบโมเดลวัดค่า MAE, MSE, RMSE
predicted2 = lm.predict(X2_test)
predicted2
```

```python
print('MAE2: ', metrics.mean_absolute_error(y2_test, predicted2))
print('MSE2: ', metrics.mean_squared_error(y2_test, predicted2))
print('RMSE2: ', np.sqrt(metrics.mean_squared_error(y2_test, predicted2)))
```

```python
#19 สร้าง distribution plot เพื่อดูว่า predicted results เป็น normal ไหม
fig = plt.figure(figsize=(10,8))
sns.distplot((y2_test-predicted2), bins=50)
```

```python
#20 ค่า RMSE ของ All-Features Multi Linear Regression มากกว่าหรือน้อย ค่า RMSE จากคู่ที่ดีที่สุดของ Simple Linear Regression มากกว่า/น้อยกว่า เท่าใด
print('ค่าRMSEของSimple LR using Avg Area Income(คู่ที่ดีที่สุด)มากกว่า ค่าRMSE ของall-feature Multi LR =', np.sqrt(metrics.mean_squared_error(y_test, predicted)) - np.sqrt(metrics.mean_squared_error(y2_test, predicted2)))
```
