```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# Data transformation
X = np.array(df['sqft_living']).reshape(-1,1)
y = np.array(df['price']).reshape(-1,1)
X
```

```python
# Data scaling
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
X
```

```python
sc_X.inverse_transform(X)
```

```python
# Split data train/test
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state=80)
```

```python
X_train
```

```python
sc_X.inverse_transform(X_train)
```

```python
#14 เทรนโมเดลแบบ Linear และ rbf ของคู่ที่ผู้เรียนคิดว่าเหมาะสม
from sklearn.svm import SVR
regressor = SVR(kernel='linear')
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.fit(X_train,y_train).predict(X_test)
```

```python
plt.scatter(sc_X.inverse_transform(X_test), sc_y.inverse_transform(y_test), color = 'red')
plt.plot(sc_X.inverse_transform(X_test), sc_y.inverse_transform(regressor.fit(X,y).predict(X_test)), color = 'blue')
plt.title('SVR')
plt.xlabel('sqrt_living')
plt.ylabel('price')
plt.show()
```

```python
y_predicted = sc_y.inverse_transform(y_pred)
y_test_inv = sc_y.inverse_transform(y_test)
y_test_inv
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test_inv, y_predicted))
print('MSE: ', metrics.mean_squared_error(y_test_inv, y_predicted))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test_inv, y_predicted)))
```

```python
data = [[metrics.mean_absolute_error(y_test_inv, y_predicted),metrics.mean_squared_error(y_test_inv, y_predicted),np.sqrt(metrics.mean_squared_error(y_test_inv, y_predicted))]]
table_data = pd.DataFrame(data,columns=['MAE','MSE','RMSE'],index=['linear'])
table_data
```

```python
#15 rbf
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X_train,y_train)
y_predicted2 = sc_y.inverse_transform(regressor.fit(X_train,y_train).predict(X_test))

fig = plt.figure(figsize=(16,8))
plt.scatter(sc_X.inverse_transform(X_test), sc_y.inverse_transform(y_test), color = 'red')
plt.plot(sc_X.inverse_transform(X_test), y_predicted2, color = 'blue')
plt.title('SVR_rbf')
plt.xlabel('sqrt_living')
plt.ylabel('price')
plt.show()
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test_inv, y_predicted2))
print('MSE: ', metrics.mean_squared_error(y_test_inv, y_predicted2))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test_inv, y_predicted2)))
```

```python
#Summary 
data = [[metrics.mean_absolute_error(y_test_inv, y_predicted2),metrics.mean_squared_error(y_test_inv, y_predicted2),np.sqrt(metrics.mean_squared_error(y_test_inv, y_predicted2))]]
table_data2 = pd.DataFrame(data,columns=['MAE','MSE','RMSE'],index=['rbf'])
table_data.append(table_data2)
```
