```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
#16 Train model แบบ Linear และ rbf (Multiple features)
df = pd.read_csv('data.csv')
```

```python
X = df[['sqft_living','sqft_above','bathrooms']]
y = np.array(df['price']).reshape(-1,1)
```

```python
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
# Split data train/test
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state=80)
```

```python
regressor = SVR(kernel='linear')
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.fit(X_train,y_train).predict(X_test)
y_predicted = sc_y.inverse_transform(y_pred)
y_test_inv = sc_y.inverse_transform(y_test)
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test_inv, y_predicted))
print('MSE: ', metrics.mean_squared_error(y_test_inv, y_predicted))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test_inv, y_predicted)))
```

```python
data = [[metrics.mean_absolute_error(y_test_inv, y_predicted),metrics.mean_squared_error(y_test_inv, y_predicted),np.sqrt(metrics.mean_squared_error(y_test_inv, y_predicted))]]
table_data = pd.DataFrame(data,columns=['MAE','MSE','RMSE'],index=['linear multifeature'])
table_data
```

```python
# เปลี่ยน kernel เป็น rbf
regressor = SVR(kernel ='rbf')
regressor.fit(X_train,y_train)
```

```python
y_predicted2 = sc_y.inverse_transform(regressor.fit(X_train,y_train).predict(X_test))
y_predicted2
```

```python
print('MAE: ', metrics.mean_absolute_error(y_test_inv, y_predicted2))
print('MSE: ', metrics.mean_squared_error(y_test_inv, y_predicted2))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test_inv, y_predicted2)))
```

```python
#Summary
data = [[metrics.mean_absolute_error(y_test_inv, y_predicted2),metrics.mean_squared_error(y_test_inv, y_predicted2),np.sqrt(metrics.mean_squared_error(y_test_inv, y_predicted2))]]
table_data2 = pd.DataFrame(data,columns=['MAE','MSE','RMSE'],index=['rbf multifeature'])
table_data.append(table_data2)
```
