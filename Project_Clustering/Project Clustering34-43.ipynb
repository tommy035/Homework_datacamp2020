```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv('College.csv')
df
```

```python
# drop outlier in Apps
df = df[df['Apps'] < 30000]
df
```

```python
# drop non-numerical data column
df = df.iloc[:,2:]
df
```

```python
#ทำ Min-Max Normalization ให้กับทุกฟีเจอร์ใน DataFrame
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()

df_scaled = min_max_scaler.fit_transform(df)
df_scaled
```

```python
df = pd.DataFrame(data=df_scaled, columns = df.columns)
df
```

```python
#34 เลือกฟีเจอร์ Apps, Accept สร้าง Hierarchical Clustering
data = df[['Apps','Accept']]
data = np.array(data)
data
```

```python
#35 ใช้ scipy สร้าง Dendrogram ให้ method = ‘ward’
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
plt.title('Dendrogram')
plt.xlabel('College')
plt.ylabel('Euc. distance')
```

```python
#36 สร้าง AgglomerativeClustering จากจำนวน Cluster ที่เหมาะสมที่สุดใน Dendrogram
# จำนวน cluster ที่เหมาะสม = 2
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=2)
y_hc = hc.fit_predict(data)
y_hc
```

```python
#37 สร้าง Clustering Visualization ของ Hierarchical Clustering
x1 = data[y_hc==0][:,0]
y1 = data[y_hc==0][:,1]
x2 = data[y_hc==1][:,0]
y2 = data[y_hc==1][:,1]

plt.figure(figsize=(10,10))
plt.scatter(x1,y1, s=20, c='b', label='cluster_1')
plt.scatter(x2,y2, s=20, c='r', label='cluster_2')
plt.title('Clustering visualization with Hierarchical C.')
plt.xlabel('Apps')
plt.ylabel('Accept')
plt.legend()
plt.grid()
```

```python
#38 เลือกฟีเจอร์ Top25perc, Outstate สร้าง Hierarchical Clustering
data = df[['Top25perc','Outstate']]
data = np.array(data)
data
```

```python
#39 ใช้ scipy สร้าง Dendrogram ให้ method = ‘ward’
plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
plt.title('Dendrogram')
plt.xlabel('College')
plt.ylabel('Eud. distance')
```

```python
#40 สร้าง AgglomerativeClustering จากจำนวน Cluster ที่เหมาะสมที่สุดใน Dendrogram
#จำนวน cluster ที่เหมาะสม = 2
hc = AgglomerativeClustering(n_clusters=2)
y_hc = hc.fit_predict(data)
y_hc
```

```python
#41 สร้าง Clustering Visualization ของ Hierarchical Clustering
x1 = data[y_hc==0][:,0]
y1 = data[y_hc==0][:,1]
x2 = data[y_hc==1][:,0]
y2 = data[y_hc==1][:,1]

plt.figure(figsize=(10,10))
plt.scatter(x1,y1, s=20, c='b', label='cluster_1')
plt.scatter(x2,y2, s=20, c='r', label='cluster_2')
plt.title('Clustering visualization with Hierarchical C.')
plt.xlabel('Top25')
plt.ylabel('Outstate')
plt.legend()
plt.grid()
```

```python
#42 เปรียบเทียบจำนวน Clusters ของ K-Means และ Hierarchical Clustering พร้อมบอกเหตุผลว่าคิดว่า Algorithm ใดเหมาะสมมากกว่าสำหรับฟีเจอร์ทั้งสองชุด
#ANSWER จำนวนclusterของK-means ดูจะให้ข้อมูล(แยกแยะข้อมูล)ได้ละเอียดกว่า ในทั้งชุด feature(Apps,Accept)และ (Top25perc,Outstate)
```

```python
#43 จากข้อ 38 ทำ Hierarchical Clustering โดยเลือกใช้จำนวน Clusters เท่ากับ K-Means และเปรียบเทียบผลลัพธ์
hc = AgglomerativeClustering(n_clusters=6)
y_hc = hc.fit_predict(data)
y_hc
```

```python
x1 = data[y_hc==0][:,0]
y1 = data[y_hc==0][:,1]
x2 = data[y_hc==1][:,0]
y2 = data[y_hc==1][:,1]
x3 = data[y_hc==2][:,0]
y3 = data[y_hc==2][:,1]
x4 = data[y_hc==3][:,0]
y4 = data[y_hc==3][:,1]
x5 = data[y_hc==4][:,0]
y5 = data[y_hc==4][:,1]
x6 = data[y_hc==5][:,0]
y6 = data[y_hc==5][:,1]

plt.figure(figsize=(10,10))
plt.scatter(x1,y1, s=20, c='blue', label='cluster_1')
plt.scatter(x2,y2, s=20, c='red', label='cluster_2')
plt.scatter(x3,y3, s=20, c='brown', label='cluster_3')
plt.scatter(x4,y4, s=20, c='gold', label='cluster_4')
plt.scatter(x5,y5, s=20, c='indigo', label='cluster_5')
plt.scatter(x6,y6, s=20, c='purple', label='cluster_6')
plt.title('Clustering visualization with Hierarchical C.')
plt.xlabel('Top25')
plt.ylabel('Outstate')
plt.legend()
plt.grid()
```
