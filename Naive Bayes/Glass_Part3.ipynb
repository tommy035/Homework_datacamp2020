```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df =  pd.read_csv('glass.csv')
df
```

# Normalization after splitting data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Type'],axis=1), df['Type'], test_size=0.3, random_state=80)
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()

min_max_scaler1 = MinMaxScaler()
X_train = min_max_scaler1.fit_transform(X_train)
X_train
```

```python
min_max_scaler2 = MinMaxScaler()
X_test = min_max_scaler2.fit_transform(X_test)
X_test
```

```python
from sklearn.naive_bayes import GaussianNB
nb4 = GaussianNB()
nb4.fit(X_train, y_train)
```

```python
pred4 = nb4.predict(X_test)
pred4
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
confusion_matrix(y_test, pred4)
```

```python
print('accuracy = {}'.format(accuracy_score(y_test, pred4)))
print('precision = {}'.format(precision_score(y_test, pred4, average='macro')))
print('recall = {}'.format(recall_score(y_test, pred4, average='macro')))
print('F1 score = {}'.format(f1_score(y_test, pred4, average='macro')))      
```
