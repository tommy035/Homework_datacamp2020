```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#84 สร้าง bar chart แสดงค่า RMSE ของทุกผลลัพธ์ทั้งหมดที่ทำมา ยกเว้น 1 indepedent variable (สองอันแรก) แล้วเปรียบเทียบว่า Model มีประสิทธิภาพมากที่สุด
x = ['LR', 'SVR','DT','RF','LR+SC', 'SVR+SC','DT+SC','RF+SC']
y = ['1121.63','1534.29','1555.94','1140.65','1121.57','1094.89','1549.17','1131.78']

fig = plt.figure(figsize=(12,8))
sns.barplot(x=x, y=y)
plt.title('RMSE Comparison')
plt.xlabel('Models')
plt.ylabel('RMSE score')

#model ที่ให้ค่าดีสุด(RMSEต่ำสุด) = SVR + SC
```

```python
#85 สร้าง bar chart แสดงค่า r2_score ของทุกผลลัพธ์ทั้งหมดที่ทำมา ยกเว้น 1 indepedent variable (สองอันแรก) แล้วเปรียบเทียบว่า Model มีประสิทธิภาพมากที่สุด
x = ['LR', 'SVR','DT','RF','LR+SC', 'SVR+SC','DT+SC','RF+SC']
y = ['0.5676','0.1910','0.1680','0.5528','0.5677','0.5880','0.1752','0.5597']

fig = plt.figure(figsize=(12,8))
sns.barplot(x=x, y=y)
plt.title('R2 Score Comparison')
plt.xlabel('Models')
plt.ylabel('R2 score')

#model ที่ให้ค่าดีสุด(R2 scoreสูงสุด)= SVR + SC
```
