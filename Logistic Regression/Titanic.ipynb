```python
### Data Cleaning
df
```

```python
df = df.drop('Cabin',axis=1) #axis = 1 :à¸¥à¸šcolumn
```

```python
df
```

```python
avg = df['Age'].mean()
```

```python
print(avg)
```

```python
df['Age'].fillna(value=avg, inplace=True)
df
```

```python
df.dropna(subset=['Embarked'], inplace = True)
df.info()
```

```python
fig = plt.figure(figsize=(12,10))
sns.heatmap(df.isnull(),cbar = False, cmap='Accent')
```

```python
df.drop(['Name','Ticket'], axis = 'columns', inplace = True)
df
```

```python
sex = pd.get_dummies(df['Sex'], drop_first=True)
sex
```

```python
embark = pd.get_dummies(df['Embarked'],drop_first=True)
embark
```

```python
df = pd.concat([df,sex,embark], axis=1)
df
```

```python
df.drop(['Sex','Embarked'], axis=1, inplace=True)
```

```python
df
```

```python
# train test model
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop(['Survived'],axis=1)
X
```

```python
y = df['Survived']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(X_train,y_train)
```

```python
predicted = logistic_regression.predict(X_test)
predicted
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
confusion_matrix(y_test, predicted)

''' Thus in binary classification, the count of true negatives is
:math:`C_{0,0}`, false negatives is :math:`C_{1,0}`, true positives is
:math:`C_{1,1}` and false positives is :math:`C_{0,1}`
'''
```

```python
print('Accuracy = ', accuracy_score(y_test, predicted))
print('F1 score = ', f1_score(y_test, predicted))
print('Precision = ', precision_score(y_test, predicted))
print('Recall score = ', recall_score(y_test, predicted))
```
