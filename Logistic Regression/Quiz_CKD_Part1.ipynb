```python
# drop columnที่หายไปเยอะ
df = df.drop(['id','rbc','rc','wc','pc','pcv'],axis=1)
df
```

```python
# fill NaN with mean values สำหรับ continuous data
column_name = ['age','bp','al','sg','su','bgr','bu','sc','sod','pot','hemo']
for i in column_name:
    avg = df[i].mean()
    df[i].fillna(value=avg, inplace=True)
```

```python
df
```

```python
df.dropna(subset=['pcc','ba','htn','dm','cad','appet','pe','ane'], inplace = True)
df
```

```python
#4 ใช้ info และ describe อธิบายข้อมูลเบื้องต้น
df.info()
```

```python
df.describe()
```

```python
df[['pcc','ba','htn','dm','cad','appet','pe','ane','classification']]
```

```python
# แปลง categorical data เป็น dummy --> พบว่ามีบางfeaturesที่raw dataเก็บข้อมูลมาไม่ดี เช่น มี tabyes(\tyes) ต้องมาcleanเพิ่ม
column_to_dummy = ['pcc','ba','htn','dm','cad','appet','pe','ane','classification']
pcc_ = pd.get_dummies(df['pcc'], drop_first=True)
ba_ = pd.get_dummies(df['ba'], drop_first=True)
htn_ = pd.get_dummies(df['htn'], drop_first=True)
dm_ = pd.get_dummies(df['dm'])
cad_ = pd.get_dummies(df['cad'])
appet_ =pd.get_dummies(df['appet'], drop_first=True)
pe_ = pd.get_dummies(df['pe'], drop_first=True)
ane_ = pd.get_dummies(df['ane'], drop_first=True)
classification_ = pd.get_dummies(df['classification'])
```

```python
df = pd.concat([df,pcc_],axis=1)
df.rename(columns={'present':'pcc_present'},inplace=True)
df.drop(['pcc'],axis=1,inplace=True)
```

```python
dm_['dm_'] = dm_[' yes']+dm_['\tyes']+dm_['yes']
dm_.drop(dm_.columns[[0,1,2,3,4]],axis=1,inplace=True)
dm_
```

```python
classification_
```

```python
classification_['ckd_'] = classification_['ckd'] + classification_['ckd\t']
classification_.drop(classification_.columns[[0,1,2,]],axis=1,inplace=True)
classification_
```

```python
df = pd.concat([df,ba_],axis=1)
df.rename(columns={'present':'ba_present'},inplace=True)
df.drop(['ba'],axis=1,inplace=True)
```

```python
df = pd.concat([df,ba_],axis=1)
df.rename(columns={'present':'ba_present'},inplace=True)
df.drop(['ba'],axis=1,inplace=True)
```

```python
df = pd.concat([df,htn_],axis=1)
df.rename(columns={'yes':'htn_'},inplace=True)
df.drop(['htn'],axis=1,inplace=True)
```

```python
df
```

```python
df = pd.concat([df,dm_],axis=1)
df.drop(['dm'],axis=1,inplace=True)
```

```python
cad_.drop(cad_.columns[[0,1]],axis=1,inplace=True)
cad_
```

```python
df= pd.concat([df,cad_],axis=1)
df.rename(columns={'yes':'cad_'},inplace=True)
df.drop(['cad'],axis=1,inplace=True)
```

```python
df= pd.concat([df,appet_],axis=1)
df.rename(columns={'poor':'poor_appet'},inplace=True)
df.drop(['appet'],axis=1,inplace=True)
```

```python
df= pd.concat([df,ane_],axis=1)
df.rename(columns={'yes':'anemia_'},inplace=True)
df.drop(['ane'],axis=1,inplace=True)
```

```python
df= pd.concat([df,pe_],axis=1)
df.rename(columns={'yes':'pedal_edema'},inplace=True)
df.drop(['pe'],axis=1,inplace=True)
```

```python
df= pd.concat([df,classification_],axis=1)
df.drop(['classification'],axis=1,inplace=True)
```
