```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
#1 โหลด csv เข้าไปใน Python Pandas
df = pd.read_csv('german_credit_data.csv')
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
#2 เขียนโค้ดแสดง หัว10แถว ท้าย10แถว และสุ่ม10แถว
df.head(10)
```

```python
df.tail(10)
```

```python
df.sample(10)
```

```python
#3 เช็คว่ามีข้อมูลที่หายไปไหม สามารถจัดการได้ตามความเหมาะสม
fig = plt.figure(figsize=(12,10))
sns.heatmap(df.isnull(),cbar=True, cmap='Greens')
```

```python
#4 ใช้ info และ describe อธิบายข้อมูลเบื้องต้น
df.info()
```

```python
df.describe()
```

```python
# จัดการ Missing values
df = df.drop(['Checking account'], axis=1)
df
```

```python
df = df.dropna(subset=['Saving accounts'])
df
```

```python
sns.heatmap(df.isnull(),cbar=True, cmap='Greens')
```

```python
#5 ใช้ pairplot ดูความสัมพันธ์เบื้องต้นของ features ที่สนใจ
sns.pairplot(data=df)
```

```python
#6 ใช้ displot เพื่อดูการกระจายของแต่ละคอลัมน์
fig = plt.figure(figsize=(10,8))
sns.distplot(df['Duration'], kde=False, bins=30)
```

```python
fig = plt.figure(figsize=(10,8))
sns.distplot(df['Credit amount'], kde=False, bins=30)
```

```python
fig = plt.figure(figsize=(10,8))
sns.distplot(df['Age'], kde=False, bins=30)
```

```python
fig = plt.figure(figsize=(10,8))
sns.distplot(df['Job'], kde=False, bins=30)
```

```python
#7 ใช้ heatmap ดูความสัมพันธ์ของคอลัมน์ที่สนใจ
sns.heatmap(df.corr(), annot=True)
```

```python
#8 สร้าง scatter plot ของความสัมพันธ์ที่มี Correlation สูงสุด
fig = plt.figure(figsize=(10,8))
sns.scatterplot(x=df['Duration'], y=df['Credit amount'])
```

```python
#9 สร้าง scatter plot ของความสัมพันธ์ที่มี Correlation ต่ำสุด
fig = plt.figure(figsize=(10,8))
sns.scatterplot(x=df['Job'], y=df['Age'])
```

```python
#10 สร้าง histogram ของ feature ที่สนใจ
df['Age'].hist(bins=30)
```

```python
#11 สร้าง box plot ของ features ที่สนใจ
fig = plt.figure(figsize=(10,8))
sns.boxplot(x=df['Saving accounts'], y=df['Credit amount'])
```

```python
fig = plt.figure(figsize=(10,8))
sns.boxplot(x=df['Job'], y=df['Credit amount'])
```
