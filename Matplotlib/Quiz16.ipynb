```python
#16 (Optional) สร้างหลายกราฟในรูปเดียว โดยมีสองคอลัมน์ คอลัมน์แรกมีสามแถว คอลัมน์สองมีสองแถว พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน โดยสมการแต่ละกราฟสามารถคิดเองได้โดยใช้ข้อมูลจากไฟล์ titanic และไม่ต้องซ้ำกัน
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('train.csv')
df.replace({'Embarked':{'S':'Southampton','C':'Cherbourg','Q':'Queenstown'}}, inplace=True)
df.dropna(subset=['Age'], inplace=True)
df.dropna(subset=['Embarked'],inplace=True)
df.replace({'Pclass':{1:'class1',2:'class2',3:'class3'}},inplace=True)
df
```

```python
a = df[df['Pclass']=='class1']['Fare'].mean()
b = df[df['Pclass']=='class2']['Fare'].mean()
c = df[df['Pclass']=='class3']['Fare'].mean()
x = [a,b,c]
PClass = [1,2,3]
```

```python
fig = plt.figure(figsize=(20,12))

ax1=fig.add_subplot(321, title = 'Number of survivors per class')
df.groupby('Pclass')['Survived'].count().plot(kind='bar',color = 'b')
ax1.set_xlabel('Class')
ax1.set_ylabel('Number of survivors')


ax2=fig.add_subplot(323,title ='Mean_fare vs Class')
ax2.bar(PClass,x)
ax2.set_xlabel('class')
ax2.set_ylabel('mean_fare')

ax3=fig.add_subplot(325,title = 'frequency of age')
ax3.hist(df['Age'],bins =10)
ax3.set_xlabel('Age')
ax3.set_ylabel('frequency')

ax4=fig.add_subplot(222,title = 'Number of each passenger class')
df.groupby('Pclass')['PassengerId'].count().plot(kind='pie')

ax5=fig.add_subplot(224,title = 'Age Vs Departure')
ax5.scatter(df['Embarked'],df['Age'],c='m')
ax5.set_xlabel('departure')
ax5.set_ylabel('age')

plt.show()

```
