```python
#1 สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7,8,9]
x = np.array(x)
y = x**2
fig = plt.figure()
plt.plot(x,y)
plt.title("Quadratic function")
plt.xlabel('X')
plt.ylabel('Y')
plt.show
```

```python
#2 สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีชมพู 
x = np.arange(0,20)
y = np.sin(x)
plt.plot(x,y,c = "pink")
plt.title('Sine Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#3 สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก
y = np.arange(0,12)
x = y**2
fig = plt.figure()
plt.plot(x,y, c='#9af764')
plt.title("X = Y^2")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

```

```python
#4 สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - - 
x = np.arange(0,20)
y = np.cos(x)
plt.plot(x,y,c = "red",ls ='--')
plt.title('Cos Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#5 สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกนใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker
x = np.arange(0,10)
y1 = x**2
y2 = x**3
fig = plt.figure()
plt.plot(x,y1, c = "red", marker = "o")
plt.plot(x,y2, c= "blue", marker = "^")
plt.title('function')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#6 สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker และเส้นสองใช้สามเหลี่ยมเป็น marker และใส่ legend และเพิ่มเส้นตารางข้างหลัง
x = np.arange(0,8)
y1 = x**2
print(x); print(y1)
y2 = x**(1/2)
print(y2)
fig= plt.figure()
plt.plot(x,y1,label='y=x^2',c='red',marker = 'o')
plt.plot(x,y2,label='x=y^2',c='blue',marker='^')
plt.title('Graphs')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=0)
plt.grid()
plt.show()
```

```python
#7 สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม 
y =np.arange(0,8)
x = y**2 + 4*y
fig= plt.figure()
plt.plot(x,y,label='x=y^2 + 4*y',c='#fef69e')
plt.title('Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([0,70])
plt.ylim([0,7])
plt.show()
```

```python
#8 สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend 
x =np.arange(0,10)
y =np.arange(0,20)
fig=plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.95,0.85])
ax1.plot(y**3,y,c='b',label='x=y^3')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('graph1')
ax1.legend(loc=0)

ax2 = fig.add_axes([0.5,0.2,0.4,0.3])
ax2.plot(x,x**2+2 ,c='m',label='y=x^2 +2')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('graph2')
ax2.legend(loc=0)

plt.show()
```
