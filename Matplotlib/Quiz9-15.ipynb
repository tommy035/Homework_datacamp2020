```python
#10 ให้หาอัตราส่วนของจุดหมายปลายทาง (Embark) และพล็อต Pie chat พร้อมทั้งให้ Queenstown เป็น explode
```

```python
df.dropna(subset=['Embarked'], inplace=True)
df
```

```python
df.info()
```

```python
df2=df.replace({'Embarked':{'S':'Southampton','C':'Cherbourg','Q':'Queenstown'}})
a = df2['Embarked'].value_counts()
A = dict(a)
Town = [k for k in A]
Number = [v for v in A.values()]
print(Town);print(Number)
```

```python
plt.figure(figsize=[5,5])
explode = [0,0,0.1]
plt.pie(Number,labels=Town,explode=explode)
plt.show()
```

```python
#11 นำราคาตั๋วโดยสารทั้งหมดมาหา histogram
F = list(df['Fare'])
plt.hist(F,bins=50)
plt.show
```

```python
#12 สร้างพล็อตแบบกระจายโดยให้อายุเป็นแกน X และ ราคาตั๋วโดยสารเป็นแกน Y
```

```python
df.dropna(subset=['Age'], inplace=True)
df
```

```python
A = df.iloc[:,5]
B = df.iloc[:,9]
Age = list(A)
Fare = list(B)
plt.figure(figsize=[5,5])
plt.scatter(Age,Fare,c='m')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
#13 สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารเฉลี่ยแต่ละชั้น
a = df[df['Pclass']==1]['Fare'].mean()
b = df[df['Pclass']==2]['Fare'].mean()
c = df[df['Pclass']==3]['Fare'].mean()
x = [a,b,c]
Class = [1,2,3]
plt.figure(figsize=[5,5])
plt.bar(Class,x)
plt.xlabel('class')
plt.ylabel('mean_Fare')
plt.show
```

```python
#14 สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารต่ออายุของผู้โดยสาร
A = df.iloc[:,5]
B = df.iloc[:,9]
Age = list(A)
Fare = list(B)
plt.figure(figsize=[5,5])
plt.bar(Age,Fare)
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
#15 สร้าง bar chart โดยใช้ข้อมูลผู้ที่นามสกุลซ้ำกัน โดยความถี่เป็นแกน Y
LN = df['Name'].str.extract( '(\w+)')
LN_duplicated = LN[LN.duplicated([0], keep=False)]
repeated_LN = LN_duplicated[0].value_counts()
repeated_LN.sort_index(0,inplace=True)
A = dict(repeated_LN)
x = [k for k in A]
y = [v for v in A.values()]
```

```python
plt.figure(figsize=[100,50])
plt.bar(x,y)
plt.xlabel('Lastname')
plt.ylabel('frequency')
plt.show()
```
