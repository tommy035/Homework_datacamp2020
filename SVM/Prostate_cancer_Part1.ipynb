```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#1 โหลด csv เข้าไปใน Python Pandas
df = pd.read_csv('prostate_cancer.csv')
df
```

```python
#2 เขียนโค้ดแสดง หัว10แถว ท้าย10แถว และสุ่ม10แถว
df.head(10)
```

```python
#3 เช็คว่ามีข้อมูลที่หายไปไหม สามารถจัดการได้ตามความเหมาะสม
fig = plt.figure(figsize=(12,10))
sns.heatmap(df.isnull(),cbar = False, cmap='Accent')
```

```python
#4 ใช้ info และ describe อธิบายข้อมูลเบื้องต้น
df.info()
```

```python
df.describe()
```

```python
# data cleansing
df['malignant'] = [1 if x == 'M' else 0 for x in df['diagnosis_result']]
df
```

```python
df.drop(['id','diagnosis_result'],axis=1,inplace=True)
df
```

```python
# save cleaned df to new csv_file
df.to_csv(r'CA_prostate.csv',index=False)
```

```python
#5 ใช้ pairplot ดูความสัมพันธ์เบื้องต้นของ features ที่สนใจ
sns.pairplot(data=df.drop(['malignant'],axis=1))
```

```python
#7 ใช้ heatmap ดูความสัมพันธ์ของคอลัมน์ที่สนใจ
fig = plt.figure(figsize=(20,10))
sns.heatmap(df.corr(), annot=True)
```

```python
#8 สร้าง scatter plot ของความสัมพันธ์ที่มี Correlation สูงสุด
fig = plt.figure(figsize=(10,8))
sns.scatterplot(x=df['area'],y=df['perimeter'],data=df)
```
