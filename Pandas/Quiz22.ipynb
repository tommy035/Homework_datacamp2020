```python
import pandas as pd
import numpy as np
df = pd.read_csv('train.csv')
df.dropna(subset=['Age'], inplace=True)
df
```

```python
lastname.sort_values(0,inplace=True)
Duplicated_lastname = lastname[lastname.duplicated()][0].unique()
Duplicated_lastname #นามสกุลที่ซ้ำกัน
```
